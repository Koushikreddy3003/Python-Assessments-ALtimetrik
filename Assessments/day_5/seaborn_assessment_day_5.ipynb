```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#checking datasets
print(sns.get_dataset_names())
```

```python
#loading data sets
car=sns.load_dataset('taxis')

#getting first 5 data of the table
print(car.head())

#geting the statistical data of the dataset
print(car.describe())

#getting the table structural data of the dataset
print(car.info())
```

```python
#getting the complete dataset/....
car
```

#Univariate Analysis

```python
sns.histplot(car['total'],bins=8, kde=True, palette='flare')
plt.title("Total Fare Distribution")
plt.xlabel("Total Fare in dollors")
plt.ylabel("Frequency")
plt.show()
```

```python
#countplot to
sns.countplot(x=car['payment'], palette='flare')
plt.title("Payment Type Distribution")
plt.xlabel("Payment Type")
plt.ylabel("No. of People Choosed")
plt.show()
```

```python
#box plot to check most travelled distances

sns.boxplot(car['distance'], palette='flare')
plt.title("Distance Distribution")
plt.xlabel("Distance in KM")
plt.ylabel("Frequency")
plt.show()
```

#bivariate analysis

```python
#1.box plot for payment

sns.boxplot(x=car['payment'], y=car['total'], palette='flare')
plt.title("Payment Type vs Total Fare")
plt.xlabel("Payment Type")
plt.ylabel("Total Fare in dollors")
plt.show()
```

```python
#2.line chart to compare distance by fare

sns.lineplot(x=car['distance'], y=car['fare'], palette='flare')
plt.title("Distance vs Fare")
plt.xlabel("Distance in KM")
plt.ylabel("Fare in dollors")
plt.show()
```

```python
#4.correlation Analysis

corr=car.corr(numeric_only=True)
print(corr)
sns.heatmap(corr, annot=True, cmap='flare')
plt.title("Correlation Analysis")
plt.show()
```

```python
#5.boxplot between distance and fare

sns.boxplot(x=car['distance'], y=car['fare'],hue=car['tip'], palette='flare')
plt.title("Distance vs Fare")
plt.xlabel("Distance in KM")
plt.ylabel("Fare in dollors")
plt.show()
```

```python
#6.outliers

Q1=car['fare'].quantile(0.25)
Q3=car['fare'].quantile(0.75)
IQR=Q3-Q1
print("IQR:",IQR)
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print("lower bound: ",lower_bound)
print("upper bound:",upper_bound)
```

```python
fare_clean= car[(car['fare']>=lower_bound) & (car['fare']<=upper_bound)]

print(f"original data size: {car.shape}")
print(f"cleaned data size: {fare_clean.shape[0]}")
```

```python
#7. Correlation after outlier

corr_clean=fare_clean.corr(numeric_only=True)
print(corr_clean)
sns.heatmap(corr_clean, annot=True, cmap='flare')
plt.title("Correlation Analysis")
plt.show()
```

```python
#getting hours from timestamp
car['pickup	hr'] = pd.to_datetime(car['pickup'])

# Extract the hour from the 'timestamp' column
car['hour'] = car['pickup	hr'].dt.hour

car
```

```python
#no.of trips going on perticular hour of 24 hours

sns.countplot(x=car['hour'], palette='flare')
plt.title("Trips per hour")
plt.xlabel("Hour")
plt.ylabel("No. of Trips")
plt.show()
```

```python
# to create new column to asign weekday or weekend

car['day_of_week'] = car['pickup'].dt.dayofweek

car['day_of_week'] = car['day_of_week'].apply(lambda x: 'Weekday' if x <= 5 else 'Weekend')
print(car)
'''car.drop(columns=['day_of_week'], inplace=True)
car'''
```

```python
#pie chart to display percentage of people travelled in weekday compared to weekend

plt.pie(car['day_of_week'].value_counts(), labels=car['day_of_week'].unique(), autopct='%1.1f%%')
plt.title("Weekday vs Weekend")
plt.legend(car['day_of_week'].unique(),loc='upper right')
plt.show()
```

```python
#plot the bar plot to no.of colour boards

sns.countplot(x=car['color'], palette='flare')
plt.title("Car color")
plt.xlabel("Count")
plt.ylabel("No. of Trips")
plt.show()
```

```python
#pairplot

sns.pairplot(car, hue='payment', palette='flare')
plt.show()
```
