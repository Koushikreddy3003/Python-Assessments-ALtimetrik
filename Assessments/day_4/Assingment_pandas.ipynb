**Q2. Create a Pandas Series from a list of student marks [65, 70, 85, 90, 55] with custom index labels
as student names.
 Calculate the average marks.
 Identify students who scored more than 75.**

```python
data={
    "ID" : [101,102,103,104,105],
    "Name":["Arun","Fadhil","Praveen","Koushik","Tharun"],
    "Marks":[65, 70, 85, 90, 55]
}

df=pd.DataFrame(data)
df
```

```python
#average marks
d=df["Marks"].mean()
print("Average marks:",d)
```

```python
#student who scored > 75
df[df["Marks"]>75]
```

1. NumPy Arrays, Indexing, Broadcasting
Q1. Create a 5x5 NumPy array of integers from 0 to 24.
 Extract the second row and third column using indexing.
 Replace all odd numbers with -1 using broadcasting.
Q2. Generate a 3D NumPy array of shape (2, 3, 4) filled with random integers between 10 and 50.
 Use slicing to extract the last 2 elements from each inner-most array.
 Broadcast a scalar value (e.g., subtract 10 from all elements).

```python
import numpy as np
```

```python
array= np.arange(25)
array
```

```python
ar=np.reshape(array,(5,5))
ar
```

```python
#Extract the second row and third column using indexing.

ext= ar[1]
print("2nd row:",ext)

ext1=ar[:,2]
print("3rd column:",ext1)
```

```python
#Replace all odd numbers with -1 using broadcasting.

odd = (ar % 2) != 0
print(odd)

# Use the mask to assign -1 to all elements that are odd
ar[odd]=-1
ar
```
